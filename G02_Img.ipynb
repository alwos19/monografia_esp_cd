```python
import numpy as np
import matplotlib.pyplot as plt
import os

input_dir = "rad13_sin_outliers/"
output_dir = "imagenes_png/"
os.makedirs(output_dir, exist_ok=True)

for archivo in os.listdir(input_dir):
    if archivo.endswith(".npy"):
        path = os.path.join(input_dir, archivo)
        array = np.load(path)

        # Normalizar
        array_norm = (array - np.min(array)) / (np.max(array) - np.min(array))

        # Guardar como .png
        nombre_png = archivo.replace(".npy", ".png")
        plt.imsave(os.path.join(output_dir, nombre_png), array_norm, cmap="gray")

        print(f"Guardado: {nombre_png}")
```
